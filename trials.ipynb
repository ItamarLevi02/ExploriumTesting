```python
# Combine all API calls into one function
def aggregate_company_data(business_id, api_key):
    """Fetch and aggregate data from all three APIs."""
    headers = {
        "accept": "application/json",
        "content-type": "application/json",
        "api_key": api_key
    }
    payload = {"business_id": business_id}

    # API Endpoints
    urls = {
        "firmographics": "https://api.explorium.ai/v1/businesses/firmographics/enrich",
        "business_challenges": "https://api.explorium.ai/v1/businesses/pc_business_challenges_10k/enrich",
        "funding_and_acquisitions": "https://api.explorium.ai/v1/businesses/funding_and_acquisition/enrich"
    }

    # Store results
    aggregated_data = {}

    for key, url in urls.items():
        response = requests.post(url, json=payload, headers=headers)
        if response.status_code == 200:
            aggregated_data[key] = response.json().get("data", {})
        else:
            aggregated_data[key] = {"error": f"API call failed: {response.status_code}"}

    # Structure the final output
    company_data = {
        "firmographics": {
            "company_name": aggregated_data["firmographics"].get("name"),
            "description": aggregated_data["firmographics"].get("business_description"),
            "location": f"{aggregated_data['firmographics'].get('country_name', 'N/A')}, {aggregated_data['firmographics'].get('region_name', 'N/A')}",
            "industry": aggregated_data["firmographics"].get("naics_description"),
            "employee_range": aggregated_data["firmographics"].get("number_of_employees_range"),
            "revenue_range": aggregated_data["firmographics"].get("yearly_revenue_range"),
            "linkedin": aggregated_data["firmographics"].get("linkedin_profile")
        },
        "business_challenges": {
            "technological_disruption": ", ".join(aggregated_data["business_challenges"].get("technological_disruption", [])),
            "security_breach": ", ".join(aggregated_data["business_challenges"].get("company_data_security_breach", [])),
            "market_saturation": ", ".join(aggregated_data["business_challenges"].get("company_market_saturation", [])),
            "competition": ", ".join(aggregated_data["business_challenges"].get("company_competition", [])),
            "customer_adoption": ", ".join(aggregated_data["business_challenges"].get("company_customer_adoption", []))
        },
        "funding_and_acquisitions": {
            "investors": aggregated_data["funding_and_acquisitions"].get("investors", []),
            "funding_rounds": aggregated_data["funding_and_acquisitions"].get("number_of_funding_rounds"),
            "last_funding_round": {
                "type": aggregated_data["funding_and_acquisitions"].get("last_funding_round_type"),
                "amount": aggregated_data["funding_and_acquisitions"].get("last_funding_round_value_usd"),
                "date": aggregated_data["funding_and_acquisitions"].get("last_funding_round_date")
            },
            "acquired_by": aggregated_data["funding_and_acquisitions"].get("acquired_by"),
            "ipo_details": {
                "ipo_size": aggregated_data["funding_and_acquisitions"].get("ipo_size_usd"),
                "ipo_date": aggregated_data["funding_and_acquisitions"].get("ipo_date")
            }
        }
    }

    return company_data

# **Test the Aggregation Function**
business_id = "8adce3ca1cef0c986b22310e369a0793"  # Replace with actual business_id
api_key = "5808751b-4634-4b0b-89d0-23a910237bf7"  # Replace with actual API key


company_info = aggregate_company_data(business_id, api_key)
print(company_info)  # Check final structured response
```

```python
from langchain.tools import BaseTool
import requests
from typing import Optional, Dict, List
from datetime import datetime, timedelta

class LinkedInPostAnalyzer(BaseTool):
    name: str = "linkedin_post_analyzer"
    description: str = "Analyzes a company's LinkedIn posts from the last 3 months to understand their recent social media presence and engagement"
    return_direct: bool = False
    
    def _run(self, business_id: str) -> str:
        api_key = "5808751b-4634-4b0b-89d0-23a910237bf7"
        url = "https://api.explorium.ai/v1/businesses/linkedin_posts/enrich"
        headers = {
            "accept": "application/json",
            "content-type": "application/json",
            "api_key": api_key
        }
        payload = {"business_id": business_id}
        
        response = requests.post(url, json=payload, headers=headers)
        
        if response.status_code == 200:
            data = response.json().get("data", [])
            
            if not data:
                return "No LinkedIn posts found for this company."
            
            # Calculate date 3 months ago
            three_months_ago = datetime.now() - timedelta(days=90)
            
            # Filter and sort posts from last 3 months
            recent_posts = [
                post for post in data 
                if datetime.strptime(post.get('created_at', ''), '%Y-%m-%dT%H:%M:%S') > three_months_ago
            ]
            
            if not recent_posts:
                return "No LinkedIn posts found from the last 3 months."
            
            # Sort posts by date (most recent first)
            recent_posts.sort(key=lambda x: x.get('created_at', ''), reverse=True)
            
            # Calculate engagement metrics for recent posts
            total_likes = sum(post.get('number_of_likes', 0) or 0 for post in recent_posts)
            total_comments = sum(post.get('number_of_comments', 0) or 0 for post in recent_posts)
            avg_likes = total_likes / len(recent_posts) if recent_posts else 0
            
            # Get the most engaging recent post
            most_engaging_post = max(
                recent_posts, 
                key=lambda x: (x.get('number_of_likes', 0) or 0) + (x.get('number_of_comments', 0) or 0)
            )
            
            # Format the output
            output = f"""
LinkedIn Activity Analysis (Last 3 Months)
----------------------------------------
Posts in Last Quarter: {len(recent_posts)}
Total Engagement: {total_likes} likes, {total_comments} comments
Average Likes per Post: {avg_likes:.1f}

Most Engaging Post:
------------------
Text: {most_engaging_post.get('post_text', 'N/A')[:200]}...
Likes: {most_engaging_post.get('number_of_likes', 'N/A')}
Comments: {most_engaging_post.get('number_of_comments', 'N/A')}
Posted: {most_engaging_post.get('created_at', 'N/A')}
URL: {most_engaging_post.get('post_url', 'N/A')}

All Posts from Last Quarter:
-------------------------"""
            
            # Add all posts from the last 3 months
            for post in recent_posts:
                post_date = datetime.strptime(post.get('created_at', ''), '%Y-%m-%dT%H:%M:%S')
                output += f"""
• Posted on: {post_date.strftime('%Y-%m-%d')}
  Engagement: {post.get('number_of_likes', 0) or 0} likes, {post.get('number_of_comments', 0) or 0} comments
  Text: {post.get('post_text', 'N/A')[:100]}...
  URL: {post.get('post_url', 'N/A')}
"""
            
            return output
        else:
            return f"Error: {response.status_code}, {response.text}"
```
